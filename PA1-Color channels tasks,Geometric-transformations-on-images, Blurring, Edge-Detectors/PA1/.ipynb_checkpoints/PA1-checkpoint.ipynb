```python
# cameraman_image = cv2.imread("Images/cameraman.jpeg")
# cameraman_image = cameraman_image[:,:,::-1]

# tx = Positive->right, Negative->left
# ty = Positive->down, Negative->up

translation_matrix = np.float32([[1, 0, -50],
                               [0, 1, 50]])

translated_image = cv2.warpAffine(src=cameraman_image, M=translation_matrix, dsize=(cameraman_image.shape[1], cameraman_image.shape[0]))

plt.imshow(translated_image, cmap="gray")
plt.show()
```

### Task 2-(b) - Rotate the image counter-clockwise by 60 degree and display the result.

```python
height, width = cameraman_image.shape[:2]
center = (width/2, height/2)

# Rotation Matrix
rotation_matrix = cv2.getRotationMatrix2D(center=center, angle=60, scale=1)

# Now rotate 60 degrees anticlockwise
rotated_image = cv2.warpAffine(src=cameraman_image, M=rotation_matrix, dsize=(width, height))

plt.imshow(rotated_image, cmap="gray")
plt.show()
```

### Task 4-(a) - Sobel Operator

```python
# To open the image with "Grayscale with only 2d shape", instead of BGR
# cameraman_image = cv2.imread("Images/cameraman.jpeg", cv2.IMREAD_GRAYSCALE)
# Define the directory name for images
image_dir = "./Images"

# Define the file names of the images you have
image_files = ["cameraman.jpeg"]

# Create dictionaries to hold the loaded images
loaded_images = {}

# Load and store the images
for image_file in image_files:
    image_path = os.path.join(image_dir, image_file)
    img = cv2.imread(image_path)

    # Convert BGR to grayscale
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Store the loaded grayscale image in the dictionary
    loaded_images[image_file.split('.')[0]] = img_gray

# Access the loaded images by their names
cameraman_image = loaded_images["cameraman"]


# Defining SOBER Kernels
horizontal_kernel = np.array([[ 1.,  2,  1],[ 0,  0,  0],[-1,-2,-1]])
vertical_kernel = np.array([[ -1.,  0,  1],[ -2,  0,  2],[-1,0,1]])

# Normalize the Kernels
kernel_sum = abs(np.sum(horizontal_kernel))
horizontal_kernel = horizontal_kernel/kernel_sum if kernel_sum else horizontal_kernel
kernel_sum = abs(np.sum(vertical_kernel))
vertical_kernel = vertical_kernel/kernel_sum if kernel_sum!=0 else vertical_kernel

# converting each pixel in large range of int-32 bits such that while calculating gradient different of one-pixel-before and ahead of current pixel, there can be more difference
cameraman_image = cameraman_image.astype('int32')

# Convolution of Image*horizontal_kernal and Image*vertical_kernel
dy_horizontal = ndimage.convolve(cameraman_image, horizontal_kernel)
dx_vertical = ndimage.convolve(cameraman_image, vertical_kernel)

# Magnitude image of above
magnitude_image = np.sqrt(dy_horizontal**2 + dx_vertical**2)
magnitude_image *= 255.0 / np.max(magnitude_image).astype(np.uint8) # Normalized

plt.figure(figsize=(15,15))
plt.subplot(1,3,1)
plt.imshow(dy_horizontal, cmap="gray")
plt.title("Horizontal Edges")

plt.subplot(1,3,2)
plt.imshow(dx_vertical, cmap="gray")
plt.title("Vertical Edges")

plt.subplot(1,3,3)
plt.imshow(magnitude_image, cmap="gray")
plt.title("Magnitude Image(Sobel)")

plt.show()
```

### Task 4-(b) - Prewitt Operator

```python
# To open the image with "Grayscale with only 2d shape", instead of BGR
# cameraman_image = cv2.imread("Images/cameraman.jpeg", cv2.IMREAD_GRAYSCALE)

# Defining PREWITT Kernels
horizontal_kernel = np.array([[ 1.,  1,  1],[ 0,  0,  0],[-1, -1, -1]])
vertical_kernel = np.array([[ 1.,  0,  -1],[ 1,  0,  -1],[1, 0, -1]])

# Normalize the Kernels
kernel_sum = abs(np.sum(horizontal_kernel))
horizontal_kernel = horizontal_kernel/kernel_sum if kernel_sum else horizontal_kernel
kernel_sum = abs(np.sum(vertical_kernel))
vertical_kernel = vertical_kernel/kernel_sum if kernel_sum!=0 else vertical_kernel

# converting each pixel in large range of int-32 bits such that while calculating gradient different of one-pixel-before and ahead of current pixel, there can be more difference
cameraman_image = cameraman_image.astype('int32')

# Convolution of Image*horizontal_kernal and Image*vertical_kernel
dy_horizontal = ndimage.convolve(cameraman_image, horizontal_kernel)
dx_vertical = ndimage.convolve(cameraman_image, vertical_kernel)

# Magnitude image of above
magnitude_image = np.sqrt(dy_horizontal**2 + dx_vertical**2)
magnitude_image *= 255.0 / np.max(magnitude_image).astype(np.uint8) # Normalized

plt.figure(figsize=(15,15))
plt.subplot(1,3,1)
plt.imshow(dy_horizontal, cmap="gray")
plt.title("Horizontal Edges")

plt.subplot(1,3,2)
plt.imshow(dx_vertical, cmap="gray")
plt.title("Vertical Edges")

plt.subplot(1,3,3)
plt.imshow(magnitude_image, cmap="gray")
plt.title("Magnitude Image(Prewitt)")

plt.show()
```

### Task 5 (20 points)

The Canny edge detector is an edge detection algorithm with multiple steps. The steps in the Canny edge detector are listed below: 

    - Smooth the image to remove the noise (Gaussian filter)
    - Find the gradients of the image
    - Apply non-maximum suppression
    - Apply double threshold. Select edge pixels by hysteresis (suppress all edges that are weak and not connected to strong edges).

Implement "Canny Edge Detector" from scratch using the steps above (for each step above, you can use library functions) and compare your result with OpenCV implementation of "Canny Edge Detector" (https://docs.opencv.org/3.4/da/d22/tutorial_py_canny.html)

```python
def gaussian_kernel(size, sigma=1):
    kernel = np.fromfunction(
        lambda x, y: (1/ (2 * np.pi * sigma ** 2)) * 
                      np.exp(-((x - (size - 1) / 2) ** 2 + (y - (size - 1) / 2) ** 2) / (2 * sigma ** 2)),
        (size, size)
    )
    return kernel / np.sum(kernel)

def apply_gaussian_blur(image, kernel_size=5, sigma=1):
    kernel = gaussian_kernel(kernel_size, sigma)
    return cv2.filter2D(image, -1, kernel)

def gradient_magnitude(dx, dy):
    return np.sqrt(dx ** 2 + dy ** 2)

def gradient_direction(dx, dy):
    return np.arctan2(dy, dx) * (180 / np.pi)

def non_maximum_suppression(gradient_magnitude, gradient_direction):
    rows, cols = gradient_magnitude.shape
    suppressed = np.zeros_like(gradient_magnitude)

    for i in range(1, rows - 1):
        for j in range(1, cols - 1):
            angle = gradient_direction[i, j]
            mag = gradient_magnitude[i, j]
            
            if (0 <= angle < 22.5) or (157.5 <= angle <= 180):
                neighbors = [gradient_magnitude[i, j - 1], gradient_magnitude[i, j + 1]]
            elif (22.5 <= angle < 67.5):
                neighbors = [gradient_magnitude[i - 1, j - 1], gradient_magnitude[i + 1, j + 1]]
            elif (67.5 <= angle < 112.5):
                neighbors = [gradient_magnitude[i - 1, j], gradient_magnitude[i + 1, j]]
            else:  # 112.5 <= angle < 157.5
                neighbors = [gradient_magnitude[i - 1, j + 1], gradient_magnitude[i + 1, j - 1]]
            
            if mag >= max(neighbors):
                suppressed[i, j] = mag

    return suppressed

def canny_edge_detection(image, low_threshold, high_threshold, kernel_size=5, sigma=1):
    # Step 1: Apply Gaussian blur to the image
    blurred_image = apply_gaussian_blur(image, kernel_size, sigma)
    
    # Step 2: Compute gradients in x and y directions
    dx = cv2.Sobel(blurred_image, cv2.CV_64F, 1, 0, ksize=3)
    dy = cv2.Sobel(blurred_image, cv2.CV_64F, 0, 1, ksize=3)
    
    # Step 3: Compute gradient magnitude and direction
    gradient_mag = gradient_magnitude(dx, dy)
    gradient_dir = gradient_direction(dx, dy)
    
    # Step 4: Perform non-maximum suppression
    suppressed = non_maximum_suppression(gradient_mag, gradient_dir)
    
    # Step 5: Apply double thresholding to detect edges
    high_threshold_value = np.max(suppressed) * high_threshold
    low_threshold_value = high_threshold_value * low_threshold
    
    edges = np.zeros_like(suppressed)
    strong_edges = (suppressed >= high_threshold_value)
    weak_edges = (suppressed >= low_threshold_value) & (suppressed < high_threshold_value)
    
    edges[strong_edges] = 255
    edges[weak_edges] = 50  # You can adjust the intensity of weak edges
    
    return edges

# Load an image
# image = cv2.imread("Images/cameraman.jpeg", 0)
image = cameraman_image.astype(np.uint8)

# Set Canny parameters
low_threshold = 0.1  # Adjust as needed
high_threshold = 0.3  # Adjust as needed

# Apply Gaussian blur
blurred_image = apply_gaussian_blur(image, kernel_size=5, sigma=1)

# Compute gradients in x and y directions
dx = cv2.Sobel(blurred_image, cv2.CV_64F, 1, 0, ksize=3)
dy = cv2.Sobel(blurred_image, cv2.CV_64F, 0, 1, ksize=3)

# Compute gradient magnitude and direction
gradient_mag = gradient_magnitude(dx, dy)
gradient_dir = gradient_direction(dx, dy)

# Perform non-maximum suppression
suppressed = non_maximum_suppression(gradient_mag, gradient_dir)

# Apply Canny edge detection
edges = canny_edge_detection(image, low_threshold, high_threshold)

# Display images in a 1x5 grid
plt.figure(figsize=(16, 4))

# Gaussian Blur Image
plt.subplot(141)
plt.imshow(blurred_image, cmap="gray")
plt.title("Gaussian Blur Image")

# Gradient Magnitude Image
plt.subplot(142)
plt.imshow(gradient_mag, cmap="gray")
plt.title("Gradient Magnitude Image")

# Non-Max Suppression Image
plt.subplot(143)
plt.imshow(suppressed, cmap="gray")
plt.title("Non-Max Suppression Image")

# Canny Edges
plt.subplot(144)
plt.imshow(edges, cmap="gray")
plt.title("Canny Edges")

plt.show()
```
